```python
# Rate per session for operated flight analysis
# (based on DEPARTURE_TIME)

# Output: table with following columns, values = ratio of operated flight

# Origin airport
# session1: 0-6am           --> 00:00 - 05:59
# session2: 6am-12noon      --> 06:00 - 11:59
# session3: 12noon-6pm      --> 12:00 - 17:59
# session4: 6pm-midnight    --> 18:00 - 23:59

# Bonus: try plotting a chart to represent
```

```python
# step1: operated flight -> filter only non-cancelled flight

# step2: count of flight in that session
# -> how to divide into 4 sessions?

# method A
# -> 4 columns, session 1, 2, 3, 4
#    each column 0/1
#    sum of the column -> count of flight in that session

# method B
# -> 1 column, session
#    can be 1/2/3/4
#    group by session, count number of flight?

# step3:  calculate ratio, divide count per session by total count of flight

# expected outcome 

#AIRPORT NAME   AIRLINE NAME   TOTAL FLIGHT  SESSION 1   SESSION 2  SESSION 3  SESSION 4 
```

```python
import numpy as np
import pandas as pd
```

```python
df_airports = pd.read_csv('airports.csv')
df_airlines = pd.read_csv('airlines.csv')
df_flights = pd.read_csv('flights.csv')
df_reports = pd.read_csv('flight_delay_report_without_index.csv')
```

```python
df_reports.sample(10)
```

```python
# filter only non-cancelled flights 
# df_flights --> column name " cancelled " = 0 

#df_flights.head(10).T 

df_NonCancelled = df_flights[df_flights["CANCELLED"]==0]
df_NonCancelled.head(10).T 
```

```python
#session 1: time 0-6am           --> 00:00 - 05:59
# session2: 6am-12noon      --> 06:00 - 11:59
# session3: 12noon-6pm      --> 12:00 - 17:59
# session4: 6pm-midnight    --> 18:00 - 23:59


df_NonCancelled['session 1'] = df_NonCancelled['DEPARTURE_TIME'].apply(lambda x:1 if 0 <= x <= 559 else 0)
df_NonCancelled['session 2'] = df_NonCancelled['DEPARTURE_TIME'].apply(lambda x:1 if 600 <= x <= 1159 else 0)
df_NonCancelled['session 3'] = df_NonCancelled['DEPARTURE_TIME'].apply(lambda x:1 if 1200 <= x <= 1759 else 0)
df_NonCancelled['session 4'] = df_NonCancelled['DEPARTURE_TIME'].apply(lambda x:1 if 1800 <= x <= 2359 else 0)
```

```python
df_departure_time =  df_NonCancelled.groupby(['AIRLINE', 'ORIGIN_AIRPORT']).agg({
    'FLIGHT_NUMBER': 'count',
    'session 1': 'sum',
    'session 2': 'sum',
    'session 3': 'sum',
    'session 4': 'sum',
})

df_departure_time.sample(10).T 
```

```python
#calculate ratio, divide count per session by total count of flight

df_departure_time['session 1 (%)'] = df_departure_time['session 1']/df_departure_time['FLIGHT_NUMBER']*100
df_departure_time['session 2 (%)'] = df_departure_time['session 2']/df_departure_time['FLIGHT_NUMBER'] *100
df_departure_time['session 3 (%)'] = df_departure_time['session 3']/df_departure_time['FLIGHT_NUMBER'] *100
df_departure_time['session 4 (%)'] = df_departure_time['session 4']/df_departure_time['FLIGHT_NUMBER'] *100


df_departure_time.sample(10)
```

```python
# rename columns name and drop unnecessary columns 
df_departure_time = df_departure_time.drop(columns=['session 1', 'session 2', 'session 3','session 4'])
df_departure_time = df_departure_time.rename(columns={'FLIGHT_NUMBER': 'total flights '})
df_departure_time
```

```python
#reset everything to columns 

df_FlightTimeSession = df_departure_time.reset_index()

df_FlightTimeSession 
```

```python
# rename Origin_airport to Airport 
#merge Airport name and Airline name 

# rename Airline_y --> Airline Name , AIRPORT --> Airport name 

df_FlightTimeSession = df_FlightTimeSession.rename(columns={'ORIGIN_AIRPORT': 'Airport'})

df_FlightTimeSession_1 = df_FlightTimeSession.merge(df_airlines, left_on = 'AIRLINE',right_on = 'IATA_CODE',how='inner')


df_FlightTimeSession_2 = df_FlightTimeSession_1.merge(df_airports, left_on = 'Airport',right_on = 'IATA_CODE',how='inner')

df_FlightTimeSession_3 = df_FlightTimeSession_2.rename(columns={'AIRLINE_y':'Airline Name','AIRPORT':'Airport Name','AIRLINE_x':'AIRLINE'})

df_FlightTimeSession_3
```

```python
# re-order the columns
df_FlightTimeSession_4= df_FlightTimeSession_3[['Airport Name','Airline Name','AIRLINE','total flights ','session 1 (%)','session 2 (%)','session 3 (%)','session 4 (%)']]
df_FlightTimeSession_4
```

```python
# merge FlightTimeSession with flight delay report 

df_flight_report = df_FlightTimeSession_4.merge(df_reports,left_on='AIRLINE',right_on='AIRLINE',how='inner')

df_flight_report.sample(10)

```

```python
# re-order the columns
df_final_flight_report= df_flight_report[['Airport Name','Airline Name','total flights ','session 1 (%)','session 2 (%)','session 3 (%)','session 4 (%)',
                                          'operated_rate', 'delayed_rate','avg_delay_min'
                                          ]]
df_final_flight_report 
```

```python
# dataframe --> csv 

df_final_flight_report.to_csv('Final_Flight_report_without_index.csv',index=False)
```
